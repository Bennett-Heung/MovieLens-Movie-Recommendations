# Exploratory Data Analysis 2

Exploratory Data Analysis using data of Genome Scores to discover:
* Correlations between genres 
* Variations in relevances of tags between genres
* Variation of relevances in ranked tags. 

```python
#Run previous notebook first
%run ./01_Data_Preparation.ipynb
```

```python
#Explode each genre from each movie's list of genres 
#And drop 'title' and 'genres' for merging with 'genome_scores'  in the next step
#Note: movies will have more than one genre, so it will not equate to the total number of movies in the dataframe 
genre = movies_df['genres'].explode()
movies_df = movies_df.merge(genre,left_index=True,right_index=True)
movies_df.columns = ['movieId','title','genres','year','genre']
movies_df.drop(['title','genres'], axis=1, inplace=True)
```

```python
#Create 'genome_genres' as 'genome_scores' is merged with each movie's genre and release year from 'movies_df'
genome_genre = genome_scores.merge(movies_df, how='left', on='movieId')
```

```python
#Standardise values for each genre - large variation between number of films released under each genre 
#So the range sum of relevances for each tag varies significantly across genre - standardising the resolve issue
genome_genres = genome_genre.groupby(['genre','tagId'])['relevance'].sum().sort_index().unstack(0)
genome_genre_scaled = preprocessing.StandardScaler().fit_transform(genome_genres.values)
genome_genre_scaled = pd.DataFrame(genome_genre_scaled)
```

```python
#Name labels are genres 
genome_genre_scaled.columns = genome_genres.columns
```

```python
#Plot correlation heatmap
plt.figure(figsize=(20,10))
sns.heatmap(genome_genre_scaled.corr(), cmap="YlGnBu", annot=True)
plt.savefig('EDA_genre_genome_corr.png')
plt.show()
```

```python
#Set up to create dataframe of each genre and their corresponding most-correlated genre  

#Store each genre into a list
most_corr_genre = []
#Store each correlation value into a list
most_corr_value = []

#For loop to find most-correlated genre and its correlation of each genre
for i in genome_genre_scaled.columns:
    most_corr_genre.append(genome_genre_scaled.corr()[i].sort_values(ascending=False).index[1])
    most_corr_value.append(genome_genre_scaled.corr()[i].sort_values(ascending=False)[1])    
```

```python
#Dataframe of each genre and their corresponding most-correlated genre
most_corr_genres = pd.DataFrame(list(zip(list(genome_genre_scaled.columns), most_corr_genre, most_corr_value)))
most_corr_genres.columns = ['genre','most_corr_genre','corr_value']

#Preview
most_corr_genres
```

```python
#Sorted above dataframe 
most_corr_genres.sort_values(by='corr_value', ascending=False)
```

```python
#The most common most-correlated genres  
most_corr_genres['most_corr_genre'].value_counts()
```

**Findings**
* The most correlated genre pairs (in order) are: Mystery and Thriller, Drama and (no genre listed), Fantasy and Adventure, and Crime and Thriller. 
* The most correlated genres are Drama, Adventure and Thriller.  

```python
#Boxplot to observe any differences in distribution in relevance values 
plt.figure(figsize=(15,7))
sns.boxplot(data = genome_genre, x = 'genre', y = 'relevance')
plt.xticks(rotation=90)
plt.savefig('EDA_genre_relevance.png')
plt.show()
```

**Findings**
* Relevances do not vary significantly across each genre - relevance of tags are approximately spread evenly across genres. 

```python
#Create 'genome_scores_sorted' as sorted 'genome_scores' by 'movieId' and 'relevance'
genome_scores_sorted = genome_scores.sort_values(by=['movieId','relevance'],ascending=[True,False])

#Preview
genome_scores_sorted
```

```python
#List of groups by 'movieId' 
#First index is index of movieId from 0 to len() 
#Second index contains dataframes of each movieId

genome_scores_grouped = list(genome_scores_sorted.groupby(['movieId']))
```

```python
#Add ranks of each tag by relevance for each 'movieId'

for i in range(len(genome_scores_grouped)): 
    #Series of ranks for each tag by relevance (without repeated ranks) 
    relevance_rank = genome_scores_grouped[i][1]['relevance'].rank(method='first',ascending=False)
    #Store ranks in new dataframe column in each movieId's dataframe as integers
    genome_scores_grouped[i][1]['relevance_rank'] = relevance_rank.astype(int)
```

```python
#Create 'genome_scores_' as the updated 'genome_scores'
genome_scores_ = []

for index, dataframe in genome_scores_grouped: 
    genome_scores_.append(dataframe)

#Concat all the dataframe movieId groups into one dataframe
genome_scores_ = pd.concat(genome_scores_, axis=0)

#Preview of updated 'genome_scores'
genome_scores_
```

```python
#Boxplot of relevances for each rank 
fig = plt.figure()
plt.figure(figsize=(15, 7))

sns.boxplot(data = genome_scores_, x = 'relevance_rank', y = 'relevance')

#Save figure
#plt.savefig('Genome_tags.png')

#Show figure
plt.show()
```

```python
#Mean and standard deviation of relevances for each rank

genome_scores_mean = genome_scores_.groupby(['relevance_rank'])['relevance'].mean()
genome_scores_index = genome_scores_mean.index
genome_scores_std = genome_scores_.groupby(['relevance_rank'])['relevance'].std()


genome_scores_mean.plot(kind='line', figsize=(15,7))
plt.fill_between(genome_scores_index, genome_scores_mean - genome_scores_std, genome_scores_mean + genome_scores_std, alpha=0.2)

plt.savefig('EDA_relevance_rank.png')
plt.show()
```

**Findings**
* The level of relevance for higher ranks of a tag is significantly higher than tags in the lower end of relevance, which are likely to not provide much information about movies.
