```python
#Get recommendations for a movie based on tag information  
def get_rec(movie_name, no_rec, movie_id_title, cosine_sim_matrix):
    """
    Provides movie recommendations from tag information.
    
    Parameters
    ----------
    movie_name: movie name 
    no_rec: number of recommendations to provide
    movie_id_title: dataframe with movieIds and their correpsonding movie titles
    cosine_sim_matrix: cosine similarity matrix
    
    Returns
    -------
    Returns a set number of movie recommendations.
    
    ---
    """
    
    #Index of the movie to locate array in cosine similarity matrix
    movie_index = movie_id_title[movie_id_title['title']==movie_name].index[0]
    
    #Array from the cosine similarity matrix 
    cosine_sim_array = cosine_sim_matrix[movie_index]

    #Create list of tuples of index 
    cosine_sim_tuples = list(enumerate(cosine_sim_array))

    #Sort tuples by cosine similarity
    cosine_sim_tuples_sorted = sorted(cosine_sim_tuples, key=lambda tup: tup[1], reverse=True)

    #List recommended movies by index and cosine similarity
    #First element is the movie name as it is the most similar to itself
    rec_index_sim = cosine_sim_tuples_sorted[1:no_rec]
    
    #Recomendation results 
    #Match movie title to correpsonding index and cosine similarity

    #Store list of movieIds and titles for each recommended movie
    rec_results_index = []
    #Store cosine similarity of each recommended movie
    rec_results_similarity = []

    #For loop for each recommended movie in 'rec_index_sim'
    for index, similarity in rec_index_sim: 
        rec_results_index.append(movie_id_title.loc[index])
        rec_results_similarity.append(similarity)

    #Dataframe of recommendations
    rec_results = pd.DataFrame(rec_results_index)
    #Add cosine similarity values to dataframe
    rec_results['similarity'] = rec_results_similarity
    
    #Final dataframe of recommendations - movieId, title, similarity (value) 
    rec_results = rec_results[['movieId','title','similarity']]
    
    return rec_results
```

## Feature Engineering

* Rank from most to least relevant tags for each movie. 
* Find the number of tags to label as relevant and for a count vectorizer, and subsequently for a cosine similarity matrix, to find similarities between movies based on the tags. 

```python
#Create 'genome_scores_sorted' as sorted 'genome_scores' by 'movieId' and 'relevance'
genome_scores_sorted = genome_scores.sort_values(by=['movieId','relevance'],ascending=[True,False])
genome_scores_sorted
```

```python
#List of groups by 'movieId' 
#First index is index of movieId from 0 to len() 
#Second index contains dataframes of each movieId

genome_scores_grouped = list(genome_scores_sorted.groupby(['movieId']))
```

```python
#Add ranks of each tag by relevance for each 'movieId'
for i in range(len(genome_scores_grouped)): 
    #Series of ranks for each tag by relevance (without repeated ranks) 
    relevance_rank = genome_scores_grouped[i][1]['relevance'].rank(method='first',ascending=False)
    #Store ranks in new dataframe column in each movieId's dataframe as integers
    genome_scores_grouped[i][1]['relevance_rank'] = relevance_rank.astype(int)
```

```python
#Create 'genome_scores_' as the updated 'genome_scores'
genome_scores_ = []
for index, dataframe in genome_scores_grouped: 
    genome_scores_.append(dataframe)

#Concat all the dataframe movieId groups into one dataframe
genome_scores_ = pd.concat(genome_scores_, axis=0)

#Preview of updated 'genome_scores'
genome_scores_
```

```python
#Boxplot of relevances for each rank 
fig = plt.figure()
plt.figure(figsize=(15, 10))

sns.boxplot(data = genome_scores_, x = 'relevance_rank', y = 'relevance')

#Show figure
plt.show()
```

The above shows the general trend of relevance decreasing for higher ranks of a tag.

This indicates that tags that are in the lower end of relevance would not provide much information to differentiate movies by the relevance of tags.

Hence, a subset of the top ranked tags by relevance is extracted to provide more information on the next steps. 

```python
#Subset of 'genome_scores_' of top 100 ranks for each movie
genome_scores_subset = genome_scores_[genome_scores_['relevance_rank'] < 101]
```

```python
#Boxplot of relevances for each rank (subset) 
fig = plt.figure()
plt.figure(figsize=(15, 10))

sns.boxplot(data = genome_scores_subset, x = 'relevance_rank', y = 'relevance')

#Save figure
plt.savefig('04_Genome_subset_tags.png')
#Show figure
plt.show()
```

Determining the number of relevant tags:  

* The range of relevances for each rank is significantly large across the board. 
* Use the median of relevances for each rank, similar to the elbow method, to find a kink in relevance. 

```python
#Subset of 'genome_scores_' of top 100 ranks for each movie
select_relevance_rank = pd.DataFrame(genome_scores_subset.groupby(['relevance_rank'])['relevance'].median())
select_relevance_rank.columns = ['median_relevance']

#Add column on the difference between the median relevances
select_relevance_rank['diff'] = select_relevance_rank.diff()
```

# Recommendations from Genome Data

The Count Vectorizer was used instead of the TF-IDF Vectorizer, as there was no need to down-weight any of the tags, especially when the twenty most relevant tags were selected from each movie.  

```python
#Define a Count Vectorizer Object, which removes all English stop words, e.g. 'a', 'the'
count_vectorizer = CountVectorizer(stop_words='english')

#Construct a Count Vectorizer matrix by fitting and transforming the tags
count_vectorizer_matrix = count_vectorizer.fit_transform(genome_scores_df['tag'])

#Shape of the Count Vectorizer matrix
count_vectorizer_matrix.shape
```

```python
#Build 'cosine_sim', the cosine similarity matrix, from the Count Vectorizer matrix
cosine_sim = cosine_similarity(count_vectorizer_matrix, count_vectorizer_matrix)

#Preview of the cosine similarity matrix
cosine_sim

#Check shape of 'cosine_sim'
print(cosine_sim.shape)
```

```python
#Get recommendations 
#Movie title is 'Up' and the number of recommendations is 20
get_rec(movie_name = 'Up', no_rec = 20, movie_id_title = genome_scores_df, cosine_sim_matrix = cosine_sim)
```
