```python
#Number of each genre 
#Note: movies will have more than one genre, so it will not equate to the total number of movies in the dataframe 
genre = movies_df['genres'].explode()
movies_df = movies_df.merge(genre,left_index=True,right_index=True)
movies_df.columns = ['movieId','title','genres','year','genre']
movies_df
```

```python
#Bar plot on the number of movies for each genre (movies data file)
movies_df['genre'].value_counts().sort_values().plot(kind='bar', figsize=(15,7))

#Save figure
plt.savefig('1EDA_genre_count.png')

#Show figure
plt.show()
```

**Findings**
* Drama and Comedy are the most common genres found, followed by Thriller, Romance and Action. 

```python
#Line plot of the number of movie releases each year (movies data file)
movies_df['year'].value_counts().sort_index().plot(kind='line', figsize=(15,7))

#Save figure
plt.savefig('2EDA_year_count.png')

#Show figure
plt.show()
```

**Findings**
* The number of movies have increased at an increasing rate over time. Note that the last dip is due to the incomplete year of 2018 in the data.

```python
#Line chart of the number of movie releases by genre
movies_df.groupby(['genre','year'])['movieId'].count().sort_index().unstack(0).plot(kind='line', figsize=(15,7))

#Save figure
plt.savefig('3EDA_genre_year_count.png')

#Show figure
plt.show()
```

**Findings**
* As well as the abundance and growth in Drama and Comedy movies, there has also been significant growth of Thriller and Documentary movies in the last 30 years. 

```python
#Merge individual genres to 'ratings_df'
ratings_df = ratings_df.merge(movies_df, how='left', on='movieId')
```

```python
#Drop irrelevant columns
ratings_df.drop(['userId','title','genres'],axis=1,inplace=True)
```

```python
#Box plot of genres in ratings data
plt.figure(figsize=(15,7))
sns.boxplot(data=ratings_df, x='genre', y='rating')
plt.xticks(rotation=90)

#Save figure
plt.savefig('4EDA_year_genre.png')

#Show figure
plt.show()
```

**Findings**
* The boxplot of each genre's distribution of ratings show higher median ratings for Drama, Crime, War, Mystery, Animation, IMAX, Film-Noir and Documentary movies, while Horror and (no genres listed) show a more likely tendency of lower ratings. The average ratings, as shown by the bar plot of average ratings for each genre, reflect these findings also. 

```python
#Bar plot of average ratings for each genre
ratings_df.groupby(['genre'])['rating'].mean().sort_values().plot(kind='bar', figsize=(15,7))

#Save figure
plt.savefig('5EDA_rating_genre.png')

#Show figure
plt.show()
```

**Findings**
* Film-Noir and War movies have the highest average rating; and Horror and '(no genres listed)' genres have the lowest average rating.

```python
#Line chart of average ratings overall and for each genre

year_mean = ratings_df.groupby(['year'])['rating'].mean()
year_std = ratings_df.groupby(['year'])['rating'].std()
year = list(year_mean.index.sort_values())

ratings_df.groupby(['genre','year'])['rating'].mean().unstack(0).plot(kind='line', figsize=(15, 7), linestyle='dashed')

year_mean.plot(kind='line', linewidth=3)
plt.fill_between(year, year_mean - year_std, year_mean + year_std, alpha=0.1)

#Save figure
plt.savefig('6EDA_rating_year_genre.png')

#Show figure
plt.show()
```

**Findings**

* An upward trend in average ratings is noticed for movies released over the earlier half the twentieth century, and have marginally declined onwards for movies released later. 

* Post 1910, there are more noticeable average ratings across genres falling under the overall average ratings, particularly between 1920 and 1960. Post 1960, average ratings across genres have converged with the marginal overall decline in average ratings.
