## Feature Engineering

Unique tags were used to filter out and duplicate tags to feature engineer tags all at once. 
Findings and actions below involve: 
* Punctuations are removed for the later process
* Foreign languages require translations - there is a chance they provide valuable information
* Tags that are purely digits remain as they are - there is a chance they provide valuable information
* One-character tags are removed - highly unlikely to hold valuable information 
* Names are converted into one word for the later process.


```python
#Find all the unique tags
unique_tags = list(set(tags_df['tag']))
unique_tags = pd.Series(unique_tags).sort_values()

#Series named 'unique_tags'
unique_tags = unique_tags.reset_index(drop=True)

#Dataframe for series
unique_tags_df = pd.DataFrame(unique_tags)
unique_tags_df.columns = ['tag']
```

### Punctuation removed

```python
#Remove punctuations
tags = []

for text in unique_tags:
    tag = str(text)
    tag_ = remove_punctuations(tag)
    tags.append(tag_)
    
unique_tags_df['removed_punctuations'] = tags
```

##### Feature Engineer Tags
* Found digit-only translations (updated tags) - kept as valuable information despite uncertain level of importance
* Remove one-character translations (updated tags) - they hold no relevant information
* Find names in tags and treat names as one word so CountVectorizer will read words appropriately. For example, two people sharing the same first name should be read as two different people rather than the same person due to their first name. 

Note: unique_tags_df1' is used here as a copy of 'unique_tags_df'

```python
#'unique_tags_df1': create copy of 'unique_tags_df'
unique_tags_df1 = unique_tags_df.copy()
```

```python
#Keep only the original tag 'tag' and the edited tags 'translations'; rest were treated as working-out columns
unique_tags_df1 = unique_tags_df1[['tag','translation']]
unique_tags_df1
```

### Digit-only Tags

```python
#Some tags are just digits - no further actions were taken due to uncertainty of significance
digits = []
for text in unique_tags_df1['translation']:
    if text.isdigit():
        digits.append(text)
        
len(digits)
```

### One-character Tags

```python
#However, translations with only one character are not relevant; will be treated as ''
one_char_tag = unique_tags_df1[unique_tags_df1['translation'].str.len()<2]
```

```python
#Store indexes of 'one-character' translations (updated tags) 
one_char_tag_index = one_char_tag.index
```

### Person Names

```python
#Find (person) names in 'unique_tags_df1' in 'translation' (updated tags) to convert them into one word 
persons_ = []
nlp = spacy.load("en_core_web_sm")

for text in unique_tags_df1['translation']:
    persons = find_persons(text)
    persons_.append(persons)
```

```python
#Check that loop to find names is done for all texts 
len(persons_)
```

```python
#Add results into 'unique_tags_df1'
unique_tags_df1['person_names'] = persons_
```

```python
#Preview of 'unique_tags_df1'
#Notice that names in 'person_names' are in lists (names are between [])
unique_tags_df1
```

```python
#Most number of names found in a text
max(unique_tags_df1['person_names'].str.len())
```

```python
#Convert names found in 'translation' (updated tags) into one word and replaced back into its updated tag
edited_tags = []

for i in unique_tags_df1.index:
    tag = unique_tags_df1['translation'][i]    
    
    if len(unique_tags_df1['person_names'][i])==3: 
        tag1_ = unique_tags_df1['person_names'][i][0]
        tag2_ = unique_tags_df1['person_names'][i][1]
        tag3_ = unique_tags_df1['person_names'][i][2]
        
        tag1 = tag1_.replace(' ','')
        tag2 = tag2_.replace(' ','')
        tag3 = tag3_.replace(' ','')
        
        edited_tag = tag.replace(tag1_,tag1).replace(tag2_,tag2).replace(tag3_,tag3)
        edited_tags.append(edited_tag)
        
    elif len(unique_tags_df1['person_names'][i])==2: 
        tag1_ = unique_tags_df1['person_names'][i][0]
        tag2_ = unique_tags_df1['person_names'][i][1]
       
        tag1 = tag1_.replace(' ','')
        tag2 = tag2_.replace(' ','')
        
        edited_tag = tag.replace(tag1_,tag1).replace(tag2_,tag2)
        edited_tags.append(edited_tag)
        
    elif len(unique_tags_df1['person_names'][i])==1: 
        tag1_ = unique_tags_df1['person_names'][i][0]
        
        tag1 = tag1_.replace(' ','')
        
        edited_tag = tag.replace(tag1_,tag1)
        edited_tags.append(edited_tag)
    else:
        edited_tags.append(tag)
```

```python
#Check if the correct number of edits to names in (updated) tags were done 
len(edited_tags)
```

```python
#Add 'edited_tags' into 'unique_tags_df1' as final edited tags to be placed into 'tags_df'
unique_tags_df1['edited_tags'] = edited_tags
unique_tags_df1
```

```python
#Merge updated tags into 'tags_df1'
tags_df1 = tags_df1.merge(unique_tags_df1[['tag','edited_tags']], how='left', on = 'tag')
```

## Recommendations by Tags
* 'combine_tags_df' is used to build recommendations.
* TF-IDF Vectorizer was used instead of a Count Vectorizer because some tags could contain words that frequently  appear and contain insignifcant value.   
* Note on a drawback that not every movie featured a tag; those movies are excluded in the list of recommendations based on tags.  

```python
#Check for NaNs
combine_tags_df['edited_tags'].isnull().sum()
```

```python
#Define a TF-IDF Vectorizer Object, which removes all English stop words, e.g. 'a', 'the'
tfidf = TfidfVectorizer(stop_words='english')

#Construct the required TF-IDF matrix by fitting and transforming the data
#TF-IDF used to 
tfidf_matrix = tfidf.fit_transform(combine_tags_df['edited_tags'])

#Output the shape of tfidf_matrix
tfidf_matrix.shape
```

```python
#Around 43k different words were used to describe around 46k movies in our dataset. 
```

```python
#Build TF-IDF (matrix) for cosine similarity matrix
tfidf_matrix = tfidf_matrix.astype(np.float32)
```

```python
#Cosine similarity matrix
#The cosine similarity scores can be calculated directly using the dot product, given the use of the TF-IDF vectorizer
#Sklearn's linear_kernel() was used instead of cosine_similarities() because it is faster.

cosine_sim = linear_kernel(tfidf_matrix,tfidf_matrix)
```
