```python
#User's movie inputs: movie titles and user ratings
user_input = [
{'title':'Breakfast Club, The', 'rating':4},
{'title':'Finding Nemo', 'rating':5},
{'title':'Toy Story', 'rating':4.5},
{'title':'Up', 'rating':4.5}] 

user_input = pd.DataFrame(user_input)
user_input
```

## Feature Engineering

* Select subset of users based on the number of movies that other users have seen that are the same as the user (given in user inputs). 

```python
#Dataframe contains dataframe of corresponding movieIds and movie titles
movie_id_title = pd.DataFrame(movies_df[['movieId','title']])
```

```python
#Filtering out the movies by title
input_movieId = movie_id_title[movie_id_title['title'].isin(user_input['title'])]
#Then merging it so we can get the movieId. It's implicitly merging it by title.
user_input = pd.merge(input_movieId, user_input)
#Preview
user_input
```

```python
#Filtering out users that have watched movies that the input has watched and storing it
subset_of_users = ratings_df[ratings_df['movieId'].isin(user_input['movieId'])]

#Preview
subset_of_users
```

```python
#Groupby creates several sub dataframes where they all have the same value in the column specified as the parameter
subset_of_users = subset_of_users.groupby(['userId'])
```

```python
#Number of users who have seen the one or more of the same movies as the user
len(subset_of_users)
```

```python
#Sorting it so users with movie most in common with the input will have priority
subset_of_users = sorted(subset_of_users, key=lambda x: len(x[1]), reverse=True)
```

```python
#Store only 

#Store userIds in 'subset_of_users'
subset_of_users_index = []
#Store dataframes of each user's userId, movieIds and ratings
subset_of_users_len = []

#For loop collecting only users who have seen all the movies from the user inputs
for user in range(len(subset_of_users)):
    if len(subset_of_users[user][1]) == len(user_input):
        subset_of_users_index.append(subset_of_users[user][0])
        subset_of_users_len.append(subset_of_users[user][1])
```

```python
#Number of users that have seen all the same movies
len(subset_of_users_index)
```

## Recommendations by (User) Ratings

1. Calculate similarities of users with Pearson's coefficient
2. Select similar users to the user (positively correlated users)
3. Calculate weighted ratings: multiply each user's Pearson's coefficient (weights) by their movie ratings 
4. Calculate weighted average recommendation scores for each movie from dividing total sum of weighted ratings by the total sum of weights
5. Sort the highest weighted average recommendation scores, followed by the number of ratings counted, as the top recommendations. 

```python
#Calculate similarities using Pearson's coefficient

#Store the user's 'userId' and corresponding Pearson coefficient
similarity_index = []

#Pearson's coefficient calculations
for user in range(len(subset_of_users_len)):  
    #Ratings from a user in the subset of users
    user_from_subset = subset_of_users_len[user]['rating'].values

    #Ratings from input user
    input_user = user_input['rating'].values

    #Pearson co-efficient
    correlation = pearsonr(user_from_subset,input_user)[0]
    similarity_index.append(correlation)
```

```python
#Create dataframe of userIds and Pearson correlations (from two lists) 
corr_matrix = pd.DataFrame(list(zip(subset_of_users_index,similarity_index)))
corr_matrix.columns = ['userId','pearson_coeff']

#Shape of 'corr_matrix'
print(len(corr_matrix))
```

```python
#Nulls are found because a user has no standard deviation in their ratings, i.e. ratings are the same for all movies  
corr_matrix[corr_matrix['pearson_coeff'].isnull()].head()
```

```python
#Keep Pearson coefficients that are not nulls
corr_matrix = corr_matrix[~corr_matrix['pearson_coeff'].isnull()]
```

```python
#Subset positively correlated users as similar users 
corr_matrix = corr_matrix[corr_matrix['pearson_coeff']>0]
corr_matrix.head()
```

```python
#Find ratings of movies from similar users 
top_users_ratings = corr_matrix.merge(ratings_df, how = 'left', on = 'userId')

#Preview
top_users_ratings 
```

```python
#Multiply similarity (pearson coefficients) of each user by their ratings
top_users_ratings['weighted_rating'] = top_users_ratings['pearson_coeff'] * top_users_ratings['rating']

#Preview 
top_users_ratings.head()
```

```python
#Sum of coefficients and weighted ratings for each movie
top_movies = top_users_ratings.groupby('movieId').sum()[['pearson_coeff','weighted_rating']]
top_movies.columns = ['sum_of_coefficients','sum_of_weighted_ratings']
top_movies.head()
```

```python
#Calculate weighted average recommendation score ('weighted_avg_recommendation_score')
top_movies['weighted_avg_recommendation_score'] = top_movies['sum_of_weighted_ratings']/top_movies['sum_of_coefficients']

#Create 'movieId' column and reset indexes 
top_movies = top_movies.reset_index()

#Create 'recommendations_df' dataframe: keeping 'movieId' and 'weighted_avg_recommendation_score' 
recommendations_df = top_movies[['movieId','weighted_avg_recommendation_score']]

#Sort 'recommendations_df' from highest to lowest 'weighted_avg_recommendation_score' 
recommendations_df = recommendations_df.sort_values(by='weighted_avg_recommendation_score',ascending=False)
```

```python
#Merge movie titles into 'recommendations_df'
recommendations_df = recommendations_df.merge(movie_id_title,how='left',on='movieId') 
recommendations_df = recommendations_df[['movieId','title','weighted_avg_recommendation_score']]
```

```python
#Top recommendations with the highest weighted average recomeendaiton score of 5
recommendations_df = recommendations_df[recommendations_df['weighted_avg_recommendation_score']==5]
```

```python
#Sort 'recommendations_df' by the number of ratings counted in the 'ratings_df' 

#Number of ratings for each movie
rating_count = ratings_df.groupby(['movieId'])['rating'].count()

#Add column of number of ratings made for each movie to 'recommendations_df' (the recommendations dataframe)
recommendations_df = recommendations_df.merge(rating_count, how = 'left', on = 'movieId')

#Sort 'recommendations_df'
recommendations_df = recommendations_df.sort_values(by='rating',ascending=False)
```
